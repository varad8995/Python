## circular linked list
list where last element contains address of first element in tail

```python
class Node:
    __slots__ = "_element","_next" ## class level member  && instance variables
    ##here we are using slots to efficiently allocate a memory and to overcome dict problems which req additional data usage
    
    
    #init function gets insitiated whenever we create an object (when we create a node it will assign it element and link)
    
    def __init__(self,element,next): 
        self._element = element #self is used to access instance variables
        self._next = next
        
class CircularLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
    def __len__(self): #return no of elements in linked list
        return self.size
    
    def isempty(self):
        return self.size == 0    
    def addlast(self,e):
        new = Node(e,None)
        if self.isempty():
            self.head = new
            new._next = new
        else:
            new._next = self.tail._next
            self.tail._next  = new
        self.tail = new
        self.size+=1
        
    def display(self):
        p = self.head
        i = 0
        while (i < len(self)):
            print(p._element,end = ' --> ')
            p = p._next
            i = i+1
        print()    
        
    def addfirst(self,e):
        new = Node(e,None)
        if self.isempty():
            self.head = new
            new._next = new
            self.tail=new
        else:
            self.tail._next = new
            new._next = self.head
        self.head = new
        self.size+=1
            
           
    def addany(self,e,position):
        new = Node(e,None)
        p = self.head
        i= 1
        while i < (position-1):
            p = p._next
            i = i+1
        new._next = p._next
        p._next = new
        self.size +=1    
        
        
    def delfirst(self):
        if self.isempty():
            print("Circular List is empty")
            return 
        e = self.head._element
        self.tail._next = self.head._next
        self.head = self.head._next
        self.size -=1
        if self.isempty():
            self.head= None
            self.tail = None
        return e    
    
    def delend(self):
        if self.isempty():
            print("Circular Linked List is empty.")
        else:
            p =  self.head
            i = 0
            while (i<len(self)-1):
                p = p._next
                i = i+1
            self.tail  = p
            p = p._next
            e=p._element
            self.size-=1
            return e
            
cl =  CircularLinkedList()    
cl.addlast(7)
print("New element is added in the circular linked list")
cl.display()
cl.addlast(6)
print("New element is added in the circular linked list")
cl.display()
cl.addlast(12)
print("New element is added in the circular linked list")
cl.display()
cl.addlast(55)
print("New element is added in the circular linked list")
cl.display()
cl.addfirst(2)
print("New element is added in the circular linked list")
cl.display()
cl.addfirst(77)
print("New element is added in the circular linked list")
cl.display()
cl.addfirst(24)
print("New element is added in the circular linked list")
cl.display()
cl.addany(46,4)
print("New element is added in the circular linked list")
cl.display()
cl.delfirst()
print("first element is deleted in the circular linked list")
cl.display()
cl.delend()
print("last element is deleted in the circular linked list")
cl.display()
print('Size :',len(cl))
```
